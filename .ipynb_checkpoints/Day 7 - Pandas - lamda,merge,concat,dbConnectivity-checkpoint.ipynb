```python
# Merge ( analogy SQL Join )
```

```python
df_emp = pd.read_csv('employees.csv')
```

```python
df_sal = pd.read_csv('salaries.csv')
```

```python
df_emp
```

```python
df_sal
```

```python
# inner join

df_emp.merge(df_sal,on='eno',how='inner')
```

```python
# left outer join

df_emp.merge(df_sal,on='eno',how='left')
```

```python
# right outer join

df_emp.merge(df_sal,on='eno',how='right')
```

```python
#  outer join
pd.merge(df_emp,df_sal,left_on='eno',right_on='eno',how='outer',suffixes=('_left','_right'))
```

```python
pd.merge(df_emp,df_sal,left_on='eno',right_on='enumber',how='left')
```

```python
pd.merge(df_emp,df_sal,left_on='eno',right_on='enumber',how='right')
```
